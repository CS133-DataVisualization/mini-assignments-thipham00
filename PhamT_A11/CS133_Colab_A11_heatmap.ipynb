## Problem 3 -  (2 points)
Your instructor's favorite video game has been Mario Kart Wii, a racing game released for the Wii platform in 2008. And, IGN agrees with her that it is a great game -- their rating for this game is a whopping 8.9! Inspired by the success of this game, your instructor is considering creating your very own racing game for the Wii platform. Perform the following analyses to help her determine which platform she should focus on.

1. Create a bar chart that shows the score for *Racing* games, for each platform. Your chart should have one bar for each platform. Provide a meaningful title to the plot.

2. Based on the bar chart, do you expect a racing game for the **Wii** platform to receive a high rating?  If not, use the pandas to find out from the dataframe `ign_scores` which gaming platform is the best for racing game?

```python
# (1 point)
# Set the width and height of the figure
# Your code here . . .
plt.figure(figsize=(10,6))
sns.barplot(y=ign_scores.index, x=ign_scores['Racing'])
plt.title('Racing Game Scores for Each Platform')
plt.ylabel('Platform')
plt.xlabel('Score')
plt.show()
```

```python
## (1 point)
## Use ign_scores to determine which gaming platform is the best
## for racing game.

## Your code here . . .
best_platform_for_racing = ign_scores['Racing'].idxmax()
print(best_platform_for_racing)
best_platform_score = ign_scores['Racing'].max()
print(best_platform_score)
```

Based on the bar chart I really don't expect a racing game for the Wii platform to receive a high rating since it clearly has the lowest score on chart. Thus to verify my assumption, I used pandas functions idxmax() to find best platform for racing and max() to find best platform score. It turns out that Xbox One is the best platform for racing with score 8.16. So my assumption about the Wii platform was correct.

#### Problem 4 - (2 points)
Since your instructor's gaming interests are pretty broad, you can help her decide to use the IGN scores to determine the choice of genre and platform.

1. Create a heatmap using the IGN scores by genre and platform and include the scores in the cells of the heatmap.
2. Base on the heatmap, which combination of genre and platform receives the highest average ratings?  Which combination receives the lowest average rankings? Write the answers in a markdown cell.

```python
## Create a heatmap
# Your code here . . .
plt.figure(figsize=(12,8))
sns.heatmap(ign_scores, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Score'})
plt.title('Heatmap of IGN Scores by Genre and Platform')
plt.xlabel('Genre')
plt.ylabel('Platform')
plt.xticks(rotation=45)
plt.show()
```

Base on the heatmap, the combination of genre and platform receives the highest average ratings is PlayStation 4-Simulation with score of 9.25

The combination receives the lowest average rankings are Game Boy Color-Fighting and Game Boy Color-Shooter with score of 4.5

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .
pokemon = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv")
filtered_data = pokemon[pokemon['Type 1'].isin(['Water', 'Normal', 'Grass', 'Bug', 'Psychic'])].set_index('Name')

numeric_data = filtered_data.select_dtypes(include=['float64', 'int64']) #select only the numeric columns for clustering

sns.clustermap(
    numeric_data,
    figsize=(12, 8),
    cmap='coolwarm',
    row_cluster=True,
    col_cluster=True,
    annot=False,
    row_colors=filtered_data['Type 1'].map({
        'Water': 'blue',
        'Normal': 'gray',
        'Grass': 'green',
        'Bug': 'orange',
        'Psychic': 'purple'
    }),
    cbar_kws={'label': 'Pokemon Stats'}
)
plt.title('Clustermap of Pokemon Stats by Type 1')
```
